# Métodos generales de simulación

## Método de Transformada inversa

* **Nota:** en este caso se parte de una distribución uniforme (0,1) continua, donde se simulara númeross pseudoaleatorios, para luego encontrar la transformada inversa de las observaciones.

```r
inversaunif<-function(n){
    U<-runif(n)
    X<-U^(1/n) #transformada de X^n
}
```

```r
x<-inversaunif(10000)
x
```

```r
hist(x, freq=FALSE)
lines(density(x), lwd=2, col="red")
```

```r
#Método aplicado a una exponencial
lambda<-2
inversaexp<-function(n){
    U<-runif(n)
    X<-(log(U)/lambda) #transformada de 1-e^lambda*x
}
```

```r
x<-inversaexp(1000)
```

```r
hist(x, freq=FALSE, xlim=c(0,4), ylim=c(0,1.5))
curve(dexp(x, lambda), add=TRUE, col="blue")
```

```r
#Contraste con Prueba Kolmogorov
ks.test(x, "pexp", lambda)
```

## Método de Aceptación-Rechazo

* **Nota:** Aqui se simula la función, apartir de que se conoce otra función determinada (U(0,1)), donde se repetira la generación aleatoria de la primera hasta que es cociente entre ambas sea menores o iguales a una constante, en donde se busca el maximo, por lo que se aplica la derivada.

```r
# 1000 observaciones de una beta (2,4)
c<-256/27
curve(dbeta(x, 2, 4))
n=1000
x<-vector()
for(i in 1:n){
    y<-runif(1)
    u<-runif(1)
    if(u<=(c*y*(1-y)^3)){
        x[i]<-y
    }else{x[i]=0}
}
x
mbetax<-x[x!=0]
mbetax
hist(mbetax, freq=FALSE)
curve(dbeta(x, 2, 4), add=TRUE, col="purple", lwd=2)
```

```r
#Contaste de Kolmogorov-Smirnov
ks.test(mbetax, "pbeta", 2, 4)
```

## Método de Composición

* **Nota:** La mixtura de f1 y f2, se define como: 𝑓(𝑥) = 𝑝𝑓1 (𝑥) + (1 − 𝑝)𝑓2 (𝑥), donde p esta entre (0,1)

**Ejemplo**

Sean 𝑓1 (𝑥) función de densidad de una 𝑁(−1, 2) y  𝑓2 (𝑥) función de densidad de una 𝑁(9, 4). Describa la mixtura resultante de  𝑓1 y 𝑓2 si 𝑝 = 0.7

*Función para simular una muestra de tamaño n de la mixtura de N(mu1, sigma1) y N(mu2, sigma2) con parametro de mezcla p*

```r
mix.norm<-function(p,mu1, sigma1, mu2, sigma2, n){
    mues.mix<-vector()
    for(i in 1:n){
        munif<-runif(1)
        if(munif<=p)mues.mix[i]<-rnorm(1, mu1, sigma1)
        else mues.mix[i]<-rnorm(1, mu2, sigma2)
    }
    mues.mix
    hist(mues.mix, freq=F)
    curve(p*dnorm(x, mu1, sigma1)+(1-p)*dnorm(x, mu2, sigma2),add=T, main="Mixtura de normales", lwd=2, col="red")
}
```

```r
mix.norm(0.7, -1,2,9,4,2000)
```

# Métodos específicos de generación de variables aleatorias discretas.

## Transformada inversa

```r
library("tidyverse")
```

```r
inv.d<-function(n=10000){
    U<-runif(n)
    X<-case_when(
        U < 0.20 ~ 1,
        U < 0.35 ~ 2,
        U < 0.60 ~3,
        TRUE ~ 4
    )
}
```

```r
siminv.d<-inv.d()
```

```r
table(siminv.d)
100*prop.table(table(siminv.d))
```

```r
#hist(siminv.d,col=palette())
qplot(siminv.d,binwidth=1,fill=I("gray"),col=I("black"))
```

```r
system.time(inv.d(1000000))
```

```r
inv.d2 <- function(n=10000){
U <- runif(n)
X <- case_when(
U < 0.40 ~ 4,
U < 0.65 ~ 3,
U < 0.85 ~ 1,
TRUE ~ 2
)
}
siminv.d2 <- inv.d2()
table(siminv.d2)
100*prop.table(table(siminv.d2))
qplot(siminv.d2,binwidth=1,fill=I("gray"),col=I("black"))
```

```r
system.time(inv.d2(1000000))
```

## Generación de variable aleatoria discreta Uniforme

* **Nota:** Un caso en el que no es necesario buscar el intervalo adecuado en el cual se encuentra el número aleatorio es cuando la variable aleatoria deseada es la variable aleatoria discreta uniforme.

```r
library(ggplot2)
```

```r
n <- 1
D.unif <- function(n,k){
floor(k*runif(n))+1
}
D.u <- D.unif(10000,30)
table(D.u)
qplot(D.u,binwidth=1,fill=I("gray"),col=I("black"))
```
