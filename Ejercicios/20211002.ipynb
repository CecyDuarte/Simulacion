## Muestreo de Gibbs

```r
Gibbs=function(n, p){
    Matriz<-matrix(ncol=2, nrow=n)
    x1<-20
    x2<-30
    Matriz[1,]<-c(x1,x2)

        for(i in 2:n)
        {
            x1=rnorm(1,p*x1, sqrt(1-(p^2)))
            x2=rnorm(1,p*x2, sqrt(1-(p^2)))
            Matriz[i,]<-c(x1,x2)
        }
    Matriz
    }
```

```r
prueba<-Gibbs(50, 0.75)
prueba
```

```r
bvn<-(Gibbs(1000,0.75))
bvn
```

```r
par(mfrow=c(1,2))
plot(bvn, col=1:1000)
plot(bvn, type="l")
```
