# Aplicaciones de la Simulación

## Método de Monte Carlo

```r
n<-1000
```

```r
x<-runif(n, -1,1)
y<-runif(n, -1,1)
```

```r
plot(x,y)
points(munif2, col="red")
curve(sqrt(1-x^2), xlim=c(-1,1), add=TRUE, col="blue")
curve(-sqrt(1-x^2), xlim=c(-1,1), add=TRUE, col="blue")
abline(v=0, lty=2)
abline(h=0, lty=2)
```

```r
munif1<-cbind(x,y) # La función cbind() pega argumentos por columnas
```

## Seleccionar los puntos que caen dentro del circulo

```r
munif2<-munif1[x^2+y^2<=1,]
```

## Estimación del número pi

```r
prob.estim<-nrow(munif2)/nrow(munif1)
```

```r
pi.estim<-4*prob.estim
```

## Hacer estimaciones para n puntos

```r
estim.pi<-function(n){
    x<-runif(n, -1,1)
    y<-runif(n, -1, 1)
    munif<-cbind(x,y)
}
```

## Varias realizaciones

```r
r<-vector()
for(i in 1:200){
    r[i]<-estim.pi(100)
}
```
